```python
import cv2
from util import *
from matplotlib import pyplot as plt

img = cv2.imread('/home/ananthu/machine_learning/face_identification_ML_model/static/uploads/test-rdj.jpg')
plt.imshow(img)
```

```python
cropped = haar_cascade(img)
plt.imshow(cropped)
```

```python
input   = stack(cropped)
plt.imshow(input.reshape(64, 32, 3))
plt.show()
```

```python
loaded_model = joblib.load('model.pkl')
predictions = loaded_model.predict(input.reshape(1, -1))
predictions
```

```python
probability = loaded_model.predict_proba(input.reshape(1, -1))
probability[0]
```

```python
output = {
        "predictions"   : predictions,
        "probabilities" : probability[0]
    }
prob = output["probabilities"]
prob[0]
```

```python
output = probability[0]

y   = 0
max = 0

for i in range(len(output)):
    
    if(output[i] > max):
        y = i
        max = output[i]

print(y,max)
print(output)
```
