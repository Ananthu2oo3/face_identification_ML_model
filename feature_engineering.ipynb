```python
import numpy as np
import pywt  #python wavelet transform
import cv2
```

```python
from matplotlib import pyplot as plt
img = cv2.imread('./data/chris_evans/chris_evans1.png')
plt.imshow(img)
```

```python
height, width, channels = img.shape
print("Height:", height)
print("Width:", width)
print("Channels:", channels)
```

```python
def w2d(height, img, mode='haar',level=1):
    
    img_array = img 
    img_array = cv2.cvtColor(img_array,cv2.COLOR_RGB2BGR)
    img_array = np.float32(img_array)
    img_array/= height

    coffs = pywt.wavedec2(img_array, mode, level)
    coffs_h = list(coffs)
    coffs_h[0] *= 0
    
    img_array_h = pywt.wavedec2(coffs_h, mode)
    img_array_h*= 255
    img_array_h = np.uint8(img_array_h)

    return img_array_h
```
